```python
import torch
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/guruiqin/deep-ml/data/california-house-price/housing_train.csv")
data = data.fillna(0)
data.head(5)
```

```python
train_data = data.drop(columns=['median_house_value'])
train_data.head(5)
```

```python
y_data = data['median_house_value']
y_data.head(5)
```

```python
# 针对数值类特征 归一化  X - X.mean() / X.std()
for col in train_data.columns:
    # print(train_data[col].dtype)
    if (train_data[col].dtype == "float64"):
        X = train_data[col]
        # print("mean===>",X.mean())
        # print("max===>",X.std())
        X = (X-X.mean())/X.std()
        train_data[col] = X
train_data.head(5)
```

```python
# one-hot 一下这里简单写下
one_hot_data = pd.get_dummies( train_data['ocean_proximity'])
train_data = train_data.drop(columns=['ocean_proximity'])
train_data = pd.concat([train_data, one_hot_data],axis=1)
train_data.head(5)
```

```python
# 如上步骤 针对值类型特征 归一化
# 针对枚举类特征  one-hot 处理了下
```

```python
train_data.to_numpy().shape[1:]
```

```python
from tensorflow import keras
model = keras.models.Sequential(
	[    
		keras.layers.Dense(30, activation="relu", input_shape=train_data.to_numpy().shape[1:]),
		keras.layers.Dense(1)
	]
)
model.summary()
```

```python
train_data.head(5)
```

```python
X_train = train_data.to_numpy()
Y_train = y_data.to_numpy()
Y_train = (Y_train - Y_train.min()) / (Y_train.max() - Y_train.min())
print(X_train.shape)
print(X_train[:1])
print(Y_train[:1])
```

```python
print(X_train[:1].shape)
model.predict(X_train[:1])
```

```python

model.compile(loss="mean_squared_error", optimizer="sgd")
# epochs 改小一点
model.fit(X_train, Y_train, epochs=5)
```
